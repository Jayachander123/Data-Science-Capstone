```python
x = [1, 4 , 6]
y = [3, 7, 10]
z = [12, 8, 4, 1]
plt.plot(x, y, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
data = pd.read_csv("countries.csv")
```

```python
data
```

```python
us = data[data.country == "United States"]
```

```python
china = data[data.country == "China"]
```

```python
us
```

```python
china
```

```python
plt.plot(us.year, us.population / 10 ** 6)
plt.plot(china.year, china.population/ 10 ** 6)
plt.legend(["this is us", "this is china"])
plt.xlabel("year")
plt.ylabel("population in millions")
```

```python
us.population / us.population.iloc[0]
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] )
plt.plot(china.year, china.population/ china.population.iloc[0])
plt.legend(["this is us", "this is china"])
plt.xlabel("year")
plt.ylabel("population growth(first year is 100)")
```
